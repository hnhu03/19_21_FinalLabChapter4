```python
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import nltk
import re
import os
import emoji
from bs4 import BeautifulSoup
from nltk.probability import FreqDist
from nltk.corpus import stopwords
src='D:/KhAI PHA WEB/19/Data1-20231102T012633Z-001/Data1'
path=os.listdir(src)
st=""
for i in path:
    label=i.split("'")[0]
    f=open('D:/KhAI PHA WEB/19/Data1-20231102T012633Z-001/Data1/'+str(label), "r", encoding="utf-8")
    text=f.read()
    st+=text 
Dataa=[]
Dataa=st.split('.')
Data=[]
for x in Dataa:
    if x.strip() != '':
        Data.append(x.strip())
vectorizer = CountVectorizer(ngram_range=(1,2))
# Tạo ra một list chứa các câu cần biểu diễn
# Biểu diễn vector đặc trưng cho các câu
features = vectorizer.fit_transform(Data)
arr=features.toarray()
print(features.toarray())
print("Vocabulary N-gram:")
j=0
for l in Data:
    print(vectorizer.fit([l]).vocabulary_)
    print(arr[j])
    j+=1
```
